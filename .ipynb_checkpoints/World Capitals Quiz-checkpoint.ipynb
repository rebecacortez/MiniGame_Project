<h3>Mini Game Project</h3>

```python
answer_A = ['A', 'a']
answer_B = ['B', 'b']
answer_C = ['C', 'c']
answer_yes = ['Y', 'y', 'yes', 'Yes']
answer_no = ['N', 'n', 'no', 'No']

advice = ('\n Use only A, B, or C \n')
score = 0

def start():
    """start the mini game"""
    print ('Welcome to World Capitals Quiz! Are you read to play? (yes/no):')
    choice = input('>>> ')
    if choice in answer_yes:
        question1()
    elif choice in answer_no:
        print('\n Thank you, bye!')


def question1(): 
    global score
    print ('\n What is the capital of Australia?')
    print ("""
    A. Sydney
    B. Canberra
    C. Melbourne""")
    choice = input('>>> ')
    if choice in answer_A:
        print ('\n Wrong answer :( \n Correct: Canberra')
    elif choice in answer_B:
        print ('\nCorrect!')
        score +=1
    elif choice in answer_C:
        print ('\n Wrong answer :( \n Correct: Canberra')
    else:
        print (advice)
    question2()

def question2(): 
    global score
    print ('\n What is the capital of Canada?')
    print ("""
    A. Ottawa
    B. Toronto
    C. Montreal""")
    choice = input('>>> ')
    if choice in answer_B:
        print ('\n Wrong answer :( \nCorrect: Ottawa')
    elif choice in answer_A:
        print ('\nCorrect!')
        score +=1
    elif choice in answer_C:
        print ('\n Wrong answer :( \nCorrect: Ottawa')
    else:
        print (advice)
    question3()
    
def question3(): 
    global score
    print ('\n Where is Caracas the capital of?')
    print ("""
    A. Venezuela
    B. El Salvador
    C. Panama""")
    choice = input('>>> ')
    if choice in answer_B:
        print ('\n Wrong answer :( \nCorrect: Venezuela')
    elif choice in answer_A:
        print ('\nCorrect!')
        score +=1
    elif choice in answer_C:
        print ('\n Wrong answer :( \nCorrect: Venezuela')
    else:
        print (advice)
    question4()

def question4(): 
    global score
    print ('\n What is the capital of Greenland?')
    print ("""
    A. Reykjavik
    B. Ilulissat
    C. Nuuk""")
    choice = input('>>> ')
    if choice in answer_B:
        print ('\n Wrong answer :( \nCorrect: Nuuk')
    elif choice in answer_C:
        print ('\nCorrect!')
        score +=1
    elif choice in answer_A:
        print ('\n Wrong answer :( \nCorrect: Nuuk')
    else:
        print (advice)
    question5()
    
def question5():
    global score
    print ('\n What is the capital of Uzbekistan?')
    print ("""
    A. Samarkand
    B. Tashkent
    C. Bukhara""")
    choice = input('>>> ')
    if choice in answer_A:
        print ('\n Wrong answer :( \nCorrect: Tashkent')
    elif choice in answer_B:
        print ('\nCorrect!')
        score += 1
    elif choice in answer_C:
        print ('\n Wrong answer :( \nCorrect: Tashkent')
    else:
        print (advice)
    bye()

def bye(): 
    """finish the game"""
    print('\n Thanks for playing this mini quiz game! Your score is:',score )
    again()

def again():
    """to play again"""
    print ('Do you want to play again? (yes/no):')
    choice = input('>>> ')
    if choice in answer_yes:
        question1()
    elif choice in answer_no:
        print('\n Thank you, bye!')
    else:
        print (advice)
start()
```
