```python
answer_yes = ['Y', 'y', 'yes', 'Yes']
answer_no = ['N', 'n', 'no', 'No']
advice = ('\n Wrong answer! Use only A, B, or C \n')
qa = {'''1. What is the capital of Australia?
    A. Sydney
    B. Canberra
    C. Melbourne''': ['B', 'b'],
    '''2. What is the capital of Canada?
    A. Ottawa
    B. Toronto
    C. Montreal''': ['A', 'a'],
    '''3. Where is Caracas the capital of?
    A. Venezuela
    B. El Salvador
    C. Panama''': ['A', 'a'],
    '''4. What is the capital of Greenland?
    A. Reykjavik
    B. Ilulissat
    C. Nuuk''': ['C', 'c'],
    '''5.What is the capital of Uzbekistan?
    A. Samarkand
    B. Tashkent
    C. Bukhara''': ['B', 'b']}
done = list(qa.keys())
def start(qa,done):
    score = 0
    choice = input('Welcome to World Capitals Quiz! Are you read to play? (yes/no): \n>>> ')
    if choice in answer_yes:
        score = question(qa,score,done)
        bye(qa,score)
    elif choice in answer_no:
        print('\n Thank you, bye!')
    return score
def question(qa,score,done):
    while len(done)>0:
        choice = input(f'{done[0]}\n')
        if choice in qa[done[0]]:
            print ('\nCorrect!')
            score +=1
        elif choice not in qa[done[0]] and choice in 'AaBbCc':
            print ('\n Wrong answer! :( ')
        else:
            print (advice)
        done.pop(0)
    return score

def bye(qa,score): 
    """finish the game"""
    print('\n Thanks for playing this mini quiz game! Your final score is:',score )
    again(qa)

def again(qa):
    """to play again"""
    score = 0
    print ('Do you want to play again? (yes/no):')
    choice = input('>>> ')
    if choice in answer_yes:
        done = list(qa.keys())
        score = question(qa,score,done)
        bye(qa,score)
    elif choice in answer_no:
        print('\n Thank you, bye!')
    else:
        print (advice)

print(f'Your final score is {start(qa,done)}')
```
